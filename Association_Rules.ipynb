```python
import pandas as pd 
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = [['Bread','Butter','Milk','Jam'],
     ['Milk','Eggs','Mayoniese','fruits','sause'],
     ['Milk','fruits','carrot','corn'],
      ['onion','sprite','maggie','sause'],
      ['maggie','Milk']]
```

```python
te = TransactionEncoder()
te_array = te.fit(df).transform(df)
te_df = pd.DataFrame(te_array,columns = te.columns_)
```

```python
te_df
```

```python
# apriori 
```

```python
frequent_itemsets = apriori(te_df,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
# Association Rules
```

```python
rules = association_rules(frequent_itemsets,metric = 'lift',min_threshold=1.1)
rules
```
