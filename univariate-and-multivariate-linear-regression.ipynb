```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import linear_model
```

Linear Regression

```python
df = pd.read_csv(r"/kaggle/input/home-prices/homeprices.csv")
df
```

```python
new_df = df.drop('price', axis = 'columns')
new_df
```

```python
price = df.price
price
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df, price)
```

```python
reg.predict([[4200]])
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
area_df = pd.read_csv(r"/kaggle/input/areas-dataset/areas.csv")
area_df.head()
```

```python
p = reg.predict(area_df)
p
```

```python
area_df['prices']=p
area_df
```

```python
area_df.to_csv("prediction")
```

#Multivariate Linear Regression

```python
df = pd.read_csv(r"/kaggle/input/home-prices/homepricesm.csv")
df
```

```python
df.bedrooms.median()
```

```python
df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df.drop('price', axis = 'columns'), df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000, 3, 40]])
```

```python
reg.predict([[2500, 4, 5]])
```
