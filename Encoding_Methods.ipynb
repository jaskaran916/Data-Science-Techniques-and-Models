```python
import pandas as pd 
import numpy as np
```

```python
data = {'Gender': ['Male','Female','Male','Male','Female'],
       'color': ['Blue','Black','Black','Blue','Blue']}
df = pd.DataFrame(data,columns = ['Gender','color'])
df
```

# One Hot Encoding

```python
df_dummies = pd.get_dummies(df,columns = ['Gender'])
```
