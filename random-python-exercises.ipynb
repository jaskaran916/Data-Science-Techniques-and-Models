Matplotlib

```python
  
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values
x = np.linspace(-10, 10, 400)  # For linear and quadratic functions
x_trig = np.linspace(-2 * np.pi, 2 * np.pi, 400)  # For trigonometric functions

# Define the equations",
y1 = -2 * x + 5          # Linear equation",
y2 = x ** 2              # Quadratic equation",
y3 = -x ** 2             # Inverted quadratic equation",
y4 = 10 * x              # Linear equation with different slope",
y5 = np.sin(x_trig)      # Sine function",
y6 = np.cos(x_trig)      # Cosine function",
  
# Create the plots",
plt.figure(figsize=(12, 8)),
   
# Plotting the linear and quadratic equations",
plt.subplot(2, 1, 1)  # Two rows, one column, first subplot
plt.plot(x, y1, label='y = -2x + 5', color='blue')
plt.plot(x, y2, label='y = x^2', color='orange'),
plt.plot(x, y3, label='y = -x^2', color='red')
plt.plot(x, y4, label='y = 10x', color='green')
plt.title('Linear and Quadratic Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid()
plt.legend()
   
# Plotting the trigonometric functions
plt.subplot(2, 1, 2)  # Two rows, one column, second subplot"
plt.plot(x_trig, y5, label='y = sin(x)', color='purple')
plt.plot(x_trig, y6, label='y = cos(x)', color='brown')
plt.title('Trigonometric Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid()
plt.legend()
  
# Show the plots
plt.tight_layout()
plt.show()
  
import matplotlib.pyplot as plt
  
# Define the x and y values
x = ['C1', 'C2', 'C3', 'C4']
y = [50, 60, 70, 85]
   
# Create a bar plot
plt.figure(figsize=(8, 5))
plt.bar(x, y, color='skyblue')
    
# Add titles and labels
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')
    
# Show grid for better readability",
plt.grid(axis='y', linestyle='--', alpha=0.7)
    
# Display the plot
plt.show()
 
```
